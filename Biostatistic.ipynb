```python
# detect potential outlies in column in Count and Rateu
```

Z = x-u/a
where z = standard score,
x = observed value
u = mean of the sample
a = standard deviation of the sample

```python
outlier_threshold = 3.5
```

```python
# detecting potential outliesrs of Count
import numpy as np
count_z_score = np.abs((df['Count']-df['Count'].mean())/df['Count'].std())
```

```python
count_outlier_mask = count_z_score > outlier_threshold
potential_count_outliers = df[count_outlier_mask]
print("potential count outliers")
print(potential_count_outliers)
u
```

```python
# detecting potential outlies of Rate
rate_z_score = np.abs((df['Rate']-df['Rate'].mean())/df['Rate'].std())

rate_outlier_mask = rate_z_score > outlier_threshold
potential_rate_outliers = df[rate_outlier_mask]
print("potential rate outliers")
print(potential_rate_outliers)

```

```python
# popultation and Rate Correlation
import pandas as pd
df = pd.read_csv('Infectious Disease 2001-2014.csv')
df.head()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

correlation_coeffiecient = df["Population"].corr(df["Rate"])
print(correlation_coeffiecient)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["Population"], df["Rate"])
plt.xlabel("Population")
plt.ylabel("Rate")
plt.title("Population vs Rate")
plt.text(0.1, 0.9, f"Correlation Coefficient: {correlation_coeffiecient:.2f}", transform=plt.gca().transAxes)
plt.show()
```

```python
 #analyzing the infected patied
 #in sex column i want male and female i.e i want to exclude total
 df = df[df['Sex'].isin (['Male','Female'])]u
```

```python
group_data = df.groupby(['Sex'])[['Count','Rate']].sum()
group_data
```

```python
# TOP 5 DISEASES FOR MALE AND FEMALE
top_diseases = {}
for sex in group_data.index:
  subset = df[df['Sex']==sex]
  subset = subset.drop_duplicates(subset=['Disease'])
  top_diseases[sex] = subset[['Disease','Rate']].nlargest(5,'Rate')
```

```python
for sex,top_disease in top_diseases.items():
  print(f"Top 5 diseases for {sex}:")
  print(top_disease)
  print()
```

```python
import matplotlib.pyplot as plt

for sex,top_disease in top_diseases.items():
  plt.figure(figsize=(10,6))
  plt.bar(top_disease['Disease'],top_disease['Rate'],color='skyblue')
  plt.xlabel('Disease')
  plt.ylabel('Rate')
  plt.title(f'Top 5 Diseases for {sex}')
```

```python
# analysis infectious diseases yearly trend
data = pd.read_csv('Infectious Disease 2001-2014.csv')
data.head()
```

```python

group_data =data.groupby(['Year','Disease'])['Count'].sum()
group_data
```

```python
top_diseases_count = group_data.groupby('Year', group_keys=False).apply(lambda x: x.nlargest(3,"Count"))
top_diseases_count
```

```python
group_data_df = group_data.reset_index()
```

```python
print(group_data_df.head())
print(group_data_df.columns)
```

```python
top_diseases_count = group_data_df.groupby('Year', group_keys=False).apply(lambda x: x.nlargest(3, 'Count'))
print(top_diseases_count)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))
sns.barplot(data=top_diseases_count, x='Year', y='Count', hue='Disease')
plt.title('Top 3 Diseases per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Disease')
plt.show()
```

```python
hitmap_data = top_diseases_count.pivot(index='Year', columns='Disease', values='Count')
plt.figure(figsize=(12, 10))
sns.heatmap(hitmap_data, annot=True, cmap='YlGnBu', fmt=".0f")
plt.title('Heatmap of Top Diseases per Year')
plt.show()
```
